```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url_main = "https://finance.naver.com/"
url_sub = "marketindex/"
page = url_main+url_sub
res = req.get(page)
soup=bs(res.content,'lxml')
```

```python
#iframe#frame_ex1
frameSrc = soup.find('iframe',id = 'frame_ex1')
print(frameSrc)
replaPage = url_main + frameSrc['src']
```

```python
print(replaPage)
```

```python
res = req.get(replaPage)
```

```python
soup = bs(res.content,'lxml')
money = soup.select('td.tit>a')
len(money)
sale = soup.select('td.sale')
len(sale)
```

```python
length = soup.select('td.tit>a')
num = []
moneys = []
sales = []

for i in range(len(length)):
    numa = i+1
    money = soup.select('td.tit>a')[i].text.strip().replace("\n","")
    sale = soup.select('td.sale')[i].text

    num.append(numa)
    moneys.append(money)
    sales.append(sale)

```

```python
moneys
```

```python
len(num), len(moneys), len(sales)
```

```python
#num,통화명, 매매기준일 값을 dataframe으로 만들기
money = pd.DataFrame({'num':num,'통화명':moneys,'매매기준 값':sales})
money.set_index('num',inplace = True)
```

```python
money
```

```python
money.to_csv('money.csv',encoding="utf-8")
```

#### 멜론 Top50위 노래제목,가수,앨범수집

```python
url = 'https://www.melon.com/chart/'
h = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36'}
res= req.get(url,headers = h)
soup = bs(res.content, 'lxml')
soup
```

```python
titles = soup.select('div.ellipsis.rank01>span',limit = 50)
#lst50 > td:nth-child(6) > div > div > div.ellipsis.rank01
singers = soup.select('div.ellipsis.rank02>a+ span',limit = 50)
albums = soup.select('div.ellipsis.rank03> a',limit = 50)
```

```python
# 1.반복문을 써서 제목 가수 앨범을 리스트에 추가
titless = []
sigerss = []
albumss =[]
length = soup.select('div.ellipsis.rank01>span >a')
for i in range(len(titles)):
    titless.append(titles[i].text.replace("\n",""))
    sigerss.append(singers[i].text)
    albumss.append(albums[i].text)
```

```python
melron = pd.DataFrame({'제목':titless,'가수':sigerss,'앨범명':albumss})
```

```python
melron.to_csv('melron.csv', encoding='utf-8')
# money.to_csv('money.csv',encoding="utf-8")
```
