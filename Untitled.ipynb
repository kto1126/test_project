```python
#main_pack > section.sc_new.cs_common_module._cs_newgame.case_normal.color_8 > div.cm_top_wrap._sticky._custom_select._header > div.title_area.type_keep._title_area > h2 > span > strong > a
```

```python
data=bs(res.content, 'lxml')
```

```python
cookie_title=data.select_one('div.title_area strong._text').text
```

```python
cookie_title
```

```python
cookie_names = data.select('div.title_box >strong > a')
cookie_names
```

```python
cookie_name_list = []
for i in cookie_names:
    cookie_name_list.append(i.text)
    print(i.text)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(cookie_name_list.columns = ['쿠키이름'])
df
```

```python
df.to_csv('cookie_names.csv', encoing = "euc-kr")
```
