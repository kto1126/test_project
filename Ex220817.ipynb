#### 2. 네이버영화 순위 제목평점 수집

```python
res = req.get("https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=20220817")
```

```python
soup = bs(res.content,'lxml')
```

```python
name = soup.select('div.tit5 > a')
len(name)
```

```python
rate = soup.select('td.point')
len(rate)
```

```python
ranks = []
names = []
ratings = []

for i in range(len(name)):
    rank=i+1
    name = soup.select('div.tit5 > a')[i].text
    rate = soup.select('td.point')[i].text
    
    ranks.append(rank)
    names.append(name)
    ratings.append(rate)
```

```python
import pandas as pd
movie_info = {'rank':ranks, 'name':names, 'rate':ratings}
movie=pd.DataFrame(movie_info)
movie.set_index('rank',inplace = True)
```

#### 3. 날짜별 영화제목, 평점수집

```python
!pip install tqdm
```

```python
from tqdm.notebook import tqdm
```

```python
url_main = 'https://movie.naver.com/movie'
url_sum = 'sdb/rank/rmovie.naver?sel=cur&tg=0&date=20220817'
```

```python
date = pd.date_range('2022-05-10',periods=100,freq='d')
#하루씩 증가d 이틀씩 증가는 2d
```

```python
#날짜, 영화제목, 평점 수집
movie_date = []
movie_name = []
movie_rate = []
#생성된 날짜만큼 반복 ==> 영화제목,평점수집
for day in tqdm(date):
    #1.영화페이지에 접근
    url_sub = 'sdb/rank/rmovie.naver?sel=cur&date=' + day.strftime('%Y%m%d')
    page = url_main + url_sub
    res = req.get(page)
    soup= bs(res.content, 'lxml')
    
    # 2. 영화제목 구하기
    length = soup.select('div.tit5 > a')
    #3. 영화 제목 평점 수집 -> 리스트에 저장
    for i in range(len(length)):
        name = soup.select('div.tit5 > a')[i].text
        rate = soup.select('td.point')[i].text
        
        movie_date.append(day)
        movie_name.append(name)
        movie_rate.append(rate)
```
