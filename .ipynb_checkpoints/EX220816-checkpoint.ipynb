#### 크롤링 : 인터넷 사이트의 페이지를 수집 해서 분류하는 것, 스크래핑 기술의 일종
#### 스크래핑 : http를 통해 내용을 긁어서 원하는 형태로 가공 
- request : 웹페이지를 가져오는 라이브러리
- bs4 : 웹페이지를 분석하는 라이브러리  

```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
res=req.get("https://news.v.daum.net/v/20220815173358873")
res
```

```python
#Http 프로토콜 status code : 요청과 응답의 상황을 표현
#200 : 성공
#404 : 페이지없음
#500 : 서버 오류
#302 : 페이지 이동
print(res.status_code)
```

```python
#실제 받아온 문서 
res.content
```

#### String, text 차이
- String 태그 하위에 문자열을 객체화, 문자열이 없으면 None반환
- text 태그하위에 있는 자식 텍스트를 문자열로 반환해줌, 문자열이 없으면 출력 없음.

```python
# 분석대상
html = """
<html>
<body>
  <h1>글자크기</h1>
  <h2>글자크기2</h2>
  <p>원하는 글자 추출</p>
  <p>크롤링 수업시간 </p>
  <p></p>
  <ul>
      <li><a href="http://www.daum.net">daum</a></li>
      <li><a href="http://www.naver.com">naver</a></li>
  </ul>
</body>
</html>  """
```

```python
soup = bs(html,"lxml")
```

```python
data=soup.find("h1")
```

```python
data.string
```

- id : 문서안에서 한번만 사용
- class : 공통적인 적용을 할 때 같은 이름으로 중복 가능

```python
html = """
<html>
<body>
  <h1 id="title">글자크기</h1>
  <p class="cstyle">원하는 글자 추출</p>
  <p id="body" align="center">크롤링 수업시간 </p>
</body>
</html>
"""
```

```python
soup=bs(html,'lxml')
```

```python
# h1태그에 id가 'title'를 지정해서 데이터 추출하기
data =soup.find('h1',id='title').text
```

```python
print(data)
```

#### css선택자 사용하기
- soup.select_one("선택자") : 선택자 요소 하나를 추출
- soup.select("선택자") : 선택자 요소 여러개를 리스트로 추출

```python
html = """
<html>
<body id="me">
<div id="smart">
      <h1 id="title">스마트 미디어 도서</h1>
      <ul class="items">
          <li>파이썬입문</li>
          <li>머신러닝</li>
          <li>딥러닝입문</li>
      </ul>
 </div>
</body>
</html>
"""
```

```python
soup=bs(html,'lxml')
```

```python
# id는 #id값 , class는 .클래스값으로 표기
data=soup.select_one("div#smart>h1").text
```

```python
data
```

```python
data= soup.select("div#smart>ul.items>li")
data
```

```python
for i in data:
    print(i.text)
```

#### 네이버 검색 페이지 가져오기

```python
keyword="쿠키런 킹덤"
url="https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query="+keyword
res = req.get(url)
res
```

```python
data=bs(res.content,'lxml')
```

```python
res=req.get("https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EC%BF%A0%ED%82%A4%EB%9F%B0+%ED%82%B9%EB%8D%A4")
soup=bs(res.content,'lxml')
data=soup.select_one("#main_pack > section.sc_new.cs_common_module._cs_newgame.case_normal.color_8 > div.cm_top_wrap._sticky._custom_select._header > div.title_area.type_keep._title_area > h2 > span > strong > a")
data.text
```

```python
cookie_names=data.select('div.title_box > strong > a')
cookie_names
```

```python
cookie_name_list=[]
for i in cookie_names:
    cookie_name_list.append(i.text)
    print(i.text)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(cookie_name_list,columns=['쿠키이름'])
df
```

```python
df.to_csv('cookie_naames.csv',encoding='euc-kr')
```
